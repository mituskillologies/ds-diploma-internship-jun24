```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([23, 34, 39, 45, 48, 55, 59, 63, 68, 78])
y = np.array([25, 29, 33, 38, 47, 51, 56, 62, 66, 75])
z = np.array([87, 83, 79, 71, 66, 59, 53, 48, 44, 31])
```

# Line Plot

```python
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.xticks(range(1, 90, 5))
plt.yticks(range(1, 90, 5))
plt.plot(x, y, marker = 'o', mfc = 'red', color = 'black', ls = '--',
        ms = 7);
plt.grid()

# markers => >, <, ^, H, o, *
```

```python
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.xticks(range(1, 90, 5))
plt.yticks(range(1, 90, 5))
plt.plot(x, z, marker = 'o', mfc = 'red', color = 'black', ls = '--',
        ms = 7);
plt.grid()
```

```python
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.xticks(range(1, 90, 5))
plt.yticks(range(1, 90, 5))
plt.plot(x, y, marker = 'o', mfc = 'red', color = 'black', ls = '--',
        ms = 7, label = '2023');
plt.plot(x, z, marker = 'o', mfc = 'yellow', color = 'black', ls = '--',
        ms = 7, label = '2024');
plt.grid()
plt.legend();
```

```python
plt.figure(figsize=(16,9))

plt.subplot(2,2,1)
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.xticks(range(1, 90, 5))
plt.yticks(range(1, 90, 5))
plt.plot(x, y, marker = 'o', mfc = 'red', color = 'black', ls = '--',
        ms = 7);
plt.grid()

plt.subplot(2,2,2)
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.xticks(range(1, 90, 5))
plt.yticks(range(1, 90, 5))
plt.plot(x, z, marker = 'o', mfc = 'red', color = 'black', ls = '--',
        ms = 7);
plt.grid()
```

# Box Plot

```python
marks = [67, 63, 57, 59, 70, 55, 68, 53, 95, 23]
```

```python
plt.boxplot(marks);
```

```python
q3 = np.percentile(marks, 75)
q1 = np.percentile(marks, 25)
q3, q1
```

```python
iqr = q3 - q1
iqr
```

```python
ub = q3 + (1.5 * iqr)
lb = q1 - (1.5 * iqr)
```

```python
ub, lb
```
